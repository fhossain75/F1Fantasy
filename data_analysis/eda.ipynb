```python
# Import libs
from main import main
from utils import calculate_personal_team

import matplotlib.pyplot as plt
```

```python
# Load team data
root_dir = "../data"

best_team_df = main(root_dir)
my_team_df = calculate_personal_team(root_dir)
```

```python
# Plot
fig, ax = plt.subplots(figsize=(15, 5))

plt.title("F1 Fantasy 2025")
plt.xlabel("Race")
plt.ylabel("Total Points")

plt.plot(best_team_df["Race"], best_team_df["Total Points"], marker='o', label="Best Team")
plt.plot(my_team_df["Race"], my_team_df["Total Points"], marker='s', color='orange', label="Faisal's Team")

# Annotate Best Teams
for x, y in zip(best_team_df["Race"], best_team_df["Total Points"]):
    ax.annotate(f"{y:.1f}", (x, y), textcoords="offset points", xytext=(0, 5), ha='center')

# Annotate Faisal's Teams
for x, y in zip(my_team_df["Race"], my_team_df["Total Points"]):
    ax.annotate(f"{y:.1f}", (x, y), textcoords="offset points", xytext=(0, 5), ha='center', color='orange')

plt.tight_layout()
plt.grid(True)
plt.legend()
plt.show()
```
